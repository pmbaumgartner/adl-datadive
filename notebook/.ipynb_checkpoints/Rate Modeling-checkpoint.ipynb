```python
num_protections = (d[['race_religion_ethnicity_enh_penalty', 'sexual_orientation_enh_penalty', 'gender_enh_penalty', 'gender_identity_enh_penalty', 'disability_enh_penalty']] == 'YES').sum(axis=1)
```

```python
enhanced_penalty_race = (d['race_religion_ethnicity_enh_penalty'] == "YES").astype(int)
```

```python
state_map = dict(zip(d['state'].unique(), range(len(d['state'].unique()))))
```

```python
year_map = dict(zip(d['date'].unique(), range(len(d['date'].unique()))))
```

```python
state_id = d['state'].map(state_map)
year_id = d['date'].map(year_map)
```

```python
with pm.Model() as race_rate_state:
    
    mu = pm.Normal('pooled_int', mu=0, sd=10)
    sd = pm.Uniform('sd', 0, 10)
    state = pm.Normal('state', mu=mu, sd=sd, shape=len(d['state'].unique()))
    
    lambda_ = (state[state_id.values])
    
    rate = pm.Poisson("obs", pm.math.exp(lambda_), observed=race_10k.values)
    
    step = pm.Metropolis()
    sample_race_state = pm.sample(10000, step=step)
```

```python
with pm.Model() as race_rate_penalty:
    
    mu = pm.Normal('pooled_int', mu=0, sd=10)
    sd = pm.Uniform('sd', 0, 10)
    state = pm.Normal('state', mu=mu, sd=sd, shape=len(d['state'].unique()))

    enh_penalty = pm.Normal('enh_penalty', mu=0, sd=10)
    
    lambda_ = (state[state_id.values] + 
               enh_penalty * enhanced_penalty_race)
    
    rate = pm.Poisson("obs", pm.math.exp(lambda_), observed=race_10k.values)
    
    step = pm.Metropolis()
    sample_race_state_penalty = pm.sample(10000, step=step)
```

```python
with pm.Model() as race_rate_datacol:
    
    mu = pm.Normal('pooled_int', mu=0, sd=10)
    sd = pm.Uniform('sd', 0, 10)
    state = pm.Normal('state', mu=mu, sd=sd, shape=len(d['state'].unique()))

    enh_penalty = pm.Normal('enh_penalty', mu=0, sd=10)
    data_collection = pm.Normal('data_collection', mu=0, sd=10)
    
    lambda_ = (state[state_id.values] + 
               data_collection * data_collection_values)
    
    rate = pm.Poisson("obs", pm.math.exp(lambda_), observed=race_10k.values)
    
    step = pm.Metropolis()
    sample_race_state_datacol = pm.sample(10000, step)
```

```python
pm.stats.waic(sample_race_state, model=race_rate_state)
```

# Gender

```python
sexual_orientation_10k = d['sexual_orientation_per_capita'] * 1000000
```

```python
enhanced_penalty_sexual_orientation = (d['sexual_orientation_enh_penalty'] == "YES").astype(int)
```

```python
with pm.Model() as gender_rate_model:
    alpha = 1 / race_10k.mean()
    
    mu = pm.Normal('pooled_int', mu=0, sd=10)
    sd = pm.Uniform('sd', 0, 10)
    state = pm.Normal('state', mu=mu, sd=sd, shape=len(d['state'].unique()))
    
    lambda_ = (state[state_id.values])

    rate = pm.Poisson("obs", pm.math.exp(lambda_), observed=gender_10k.values)
    
    #obs = pm.Deterministic('rate', rate)
    
    step = pm.Metropolis()
    sample_state_gender = pm.sample(10000, step)
```

```python
pm.stats.dic(sample_state_gender, model=gender_rate_model)
```

```python
with pm.Model() as gender_rate_penalty:
    alpha = 1 / race_10k.mean()
    
    mu = pm.Normal('pooled_int', mu=0, sd=10)
    sd = pm.Uniform('sd', 0, 10)
    state = pm.Normal('state', mu=mu, sd=sd, shape=len(d['state'].unique()))
    
    enh_penalty = pm.Normal('enh_penalty', mu=0, sd=10)
    n_protections = pm.Normal('n_protections', mu=0, sd=10)
    
    lambda_ = (state[state_id.values] + enh_penalty * enhanced_penalty_sexual_orientation + n_protections * num_protections)
    
    rate = pm.Poisson("obs", pm.math.exp(lambda_), observed=sexual_orientation_10k.values)
    
    #obs = pm.Deterministic('rate', rate)
    
    sample_state_penalty_gender = pm.sample(10000)
```
