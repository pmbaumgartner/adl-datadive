```python
with pm.Model() as race_rate_state:
    
    mu = pm.Normal('pooled_int', mu=0, sd=10)
    sd = pm.Uniform('sd', 0, 10)
    state = pm.Normal('state', mu=mu, sd=sd, shape=len(d['state'].unique()))
    
    lambda_ = (state[state_id.values])
    
    rate = pm.Poisson("obs", pm.math.exp(lambda_), observed=race_1m.values)
    
    sample_race_state = pm.sample(100000, step=pm.Metropolis(), start=pm.find_MAP())
```

```python
with pm.Model() as race_rate_penalty:
    
    mu = pm.Normal('pooled_int', mu=0, sd=10)
    sd = pm.Uniform('sd', 0, 10)
    state = pm.Normal('state', mu=mu, sd=sd, shape=len(d['state'].unique()))

    enh_penalty = pm.Normal('enh_penalty', mu=0, sd=10, shape=2)
    
    lambda_ = (state[state_id.values] + 
               enh_penalty[enhanced_penalty_race.values])
    
    rate = pm.Poisson("obs", pm.math.exp(lambda_), observed=race_1m.values)
    
    sample_race_state_penalty = pm.sample(100000, step=pm.Metropolis(), start=pm.find_MAP())
```

```python
with pm.Model() as race_rate_datacol:
    
    mu = pm.Normal('pooled_int', mu=0, sd=10)
    sd = pm.Uniform('sd', 0, 10)
    state = pm.Normal('state', mu=mu, sd=sd, shape=len(d['state'].unique()))

    data_collection = pm.Normal('data_collection', mu=0, sd=10, shape=2)
    
    lambda_ = (state[state_id.values] + 
               data_collection[data_collection_values])
    
    rate = pm.Poisson("obs", pm.math.exp(lambda_), observed=race_1m.values)
    
    sample_race_state_datacol = pm.sample(100000, step=pm.Metropolis(), start=pm.find_MAP())
```

```python
with pm.Model() as race_rate_datacol_penalty:
    
    mu = pm.Normal('pooled_int', mu=0, sd=10)
    sd = pm.Uniform('sd', 0, 10)
    state = pm.Normal('state', mu=mu, sd=sd, shape=len(d['state'].unique()))
    
    data_collection = pm.Normal('data_collection', mu=0, sd=10, shape=2)
    enh_penalty = pm.Normal('enhanced_penalty', shape=2)
    
    lambda_ = (state[state_id.values] +
               data_collection[data_collection_values] +
              enh_penalty[enhanced_penalty_race.values])
    
    rate = pm.Poisson("obs", pm.math.exp(lambda_), observed=race_1m.values)
        
    sample_race_rate_datacol_penalty = pm.sample(100000, step=pm.Metropolis(), start=pm.find_MAP())
```
